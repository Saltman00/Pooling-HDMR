```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt


(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

image = x_test[0]  

image = tf.expand_dims(image, axis=0)  # (1, 32, 32, 3)
image = tf.cast(image, tf.float32) / 255.0  
conv_layer = tf.keras.layers.Conv2D(filters=16, kernel_size=(3, 3), activation='relu', padding="same")

feature_maps = conv_layer(image).numpy()[0]  

fig, axes = plt.subplots(4, 4, figsize=(12, 12))
axes = axes.flatten()  

for i in range(16):  
    axes[i].imshow(feature_maps[:, :, i])  
    axes[i].axis("off")

```

```python
from skimage.util import random_noise
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
image2=x_test[0]  
noisy_image=random_noise(image2,mode='s&p',amount=0.1)
noisy_image=tf.expand_dims(noisy_image,axis=0)
feature_maps2 = conv_layer(noisy_image).numpy()[0]  
fig2, axes2 = plt.subplots(4, 4, figsize=(12, 12))
axes2 = axes2.flatten()  

for i in range(16):  
    axes2[i].imshow(feature_maps2[:, :, i], cmap="gray")  
    axes2[i].axis("off")
```
