```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# CIFAR-10 veri setini yükle
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# İlk görüntüyü seç ve NumPy'ye çevir (TensorFlow tensör olabilir!)
image = np.array(x_test[0])  # (32,32,3)

# Max Pooling sonucu için 16x16x3 boş bir matris oluştur
pooled_image = np.zeros((16, 16, 3), dtype=np.float32)

# 2x2 bloklar üzerinden ilerle ve her bloğun maksimum değerini al
for i in range(16):
    for j in range(16):
        temp_mat = image[i*2:i*2+2, j*2:j*2+2, :]  # 2x2 blok al
        pooled_image[i, j, :] = np.max(temp_mat, axis=(0, 1))  # Max değeri al

# Görüntüyü uint8 formatına çevir
pooled_image = pooled_image.astype(np.uint8)
pool_layer=tf.keras.layers.MaxPooling2D(pool_size=(2,2),strides=2)
im2=tf.expand_dims(image, axis=0)
pooled_image2=pool_layer(im2)
pooled_image2=np.array(pooled_image2[0])

# Görüntüleri karşılaştır
plt.figure(figsize=(8, 4))

plt.subplot(2, 2, 1)
plt.imshow(image)
plt.title("Orijinal (32x32)")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(pooled_image)
plt.title("Max Pooling by hand (16x16)")
plt.axis("off")

plt.subplot(2,2, 3)
plt.imshow(pooled_image2)
plt.title("Max Pooling  by alghortim(16x16)")
plt.axis("off")

plt.show()
```
